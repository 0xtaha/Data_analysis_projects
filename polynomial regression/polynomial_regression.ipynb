```python
import pandas as pd 
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['Var_X'],data['Var_Y'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = data['Var_X'].values.reshape(-1,1)
y = data['Var_Y'].values
```

```python
poly_fit = PolynomialFeatures(degree= 4)
X_poly = poly_fit.fit_transform(x)
```

```python
poly_model = LinearRegression(fit_intercept = False).fit(X_poly, y)
```

```python
predected_y = poly_model.predict(X_poly)
```

```python
plt.scatter(x,y)
plt.scatter(x,predected_y, color = 'red')

plt.show()
```
